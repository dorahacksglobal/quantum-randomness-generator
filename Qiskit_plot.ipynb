```python
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import DensityMatrix
from qiskit.visualization import plot_state_city
```

```python
# x0=sigma_z, y0=1/sqrt(2)*(sigma_z+sigma_x)
qc00 = QuantumCircuit(2,2)
qc00.h(0)
qc00.cnot(0,1)
qc00.id(0)
qc00.s(1)
qc00.h(1)
qc00.t(1)
qc00.h(1)
qc00.measure([0, 1], [0, 1])
# x0=sigma_z, y1=1/sqrt(2)*(sigma_z-sigma_x)
qc01 = QuantumCircuit(2,2)
qc01.h(0)
qc01.cnot(0,1)
qc01.id(0)
qc01.s(1)
qc01.h(1)
qc01.tdg(1)
qc01.h(1)
qc01.measure([0, 1], [0, 1])
# x1=sigma_x, y0=1/sqrt(2)*(sigma_z+sigma_x)
qc10 = QuantumCircuit(2,2)
qc10.h(0)
qc10.cnot(0,1)
qc10.h(0)
qc10.s(1)
qc10.h(1)
qc10.t(1)
qc10.h(1)
qc10.measure([0, 1], [0, 1])
# x0=sigma_x, y1=1/sqrt(2)*(sigma_z-sigma_x)
qc11 = QuantumCircuit(2,2)
qc11.h(0)
qc11.cnot(0,1)
qc11.h(0)
qc11.s(1)
qc11.h(1)
qc11.tdg(1)
qc11.h(1)
qc11.measure([0, 1], [0, 1])
```

```python
qc00.draw(output='mpl', initial_state=True)
```
